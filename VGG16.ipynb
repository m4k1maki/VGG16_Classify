```python
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator # Changed import statement to include tensorflow
import numpy as np
```

```python
trdata = ImageDataGenerator()
traindata = trdata.flow_from_directory(directory="/content/drive/MyDrive/Colab Notebooks/cats_and_dogs_filtered/train",target_size=(224,224),batch_size=16)
tsdata = ImageDataGenerator()
testdata = tsdata.flow_from_directory(directory="/content/drive/MyDrive/Colab Notebooks/cats_and_dogs_filtered/validation", target_size=(224,224),batch_size = 16)
```

```python
model = Sequential()
model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=2, activation="softmax"))

```

```python
from keras.optimizers import Adam
# use learning_rate instead of lr
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
# use save_freq instead of period and add the .keras extension to the filepath
checkpoint = ModelCheckpoint("cat_dog_fine_tune_model.h5", monitor='val_acc', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', save_freq=1)
# Set mode to 'max' to tell EarlyStopping to maximize val_acc
early = EarlyStopping(monitor='val_acc', min_delta=0, patience=20, verbose=1, mode='max')
```

```python
hist = model.fit(traindata,
                    steps_per_epoch=100,
                    validation_data= testdata,
                    validation_steps=10,
                    epochs=200,
                    callbacks=[checkpoint,early])
```

```python
model.save('/Colab Notebooks/cat_dog_fine_tune_model.h5')
```

```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2

from keras.preprocessing import image
path = "/content/drive/MyDrive/Colab Notebooks/cats_and_dogs_filtered/validation/dogs/dog.2022.jpg"
img = image.load_img(path,target_size=(224,224))
img = np.asarray(img)
img = np.expand_dims(img, axis=0)
from keras.models import load_model
saved_model = load_model("/content/drive/MyDrive/Colab Notebooks/cat_dog_fine_tune_model.h5")
img1 = cv2.imread(path)
output = saved_model.predict(img)
cv2_imshow(img1)
if output[0][0] > output[0][1]:
    print("cat")
else:
    print('dog')

```

```python
loss, accuracy = saved_model.evaluate(testdata)
print('Độ chính xác:', accuracy)
```
